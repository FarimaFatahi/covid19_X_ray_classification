**Majority voting**

```python
FILES_PATH = GOOGLE_DRIVE_PATH + 'Model_LungSegmentation/'
test_csv = os.path.join(GOOGLE_DRIVE_PATH,'Label_Files/test_labels.csv')
inceptionV3_test_results = loadmat(FILES_PATH + 'InceptionV3_Test_Index.mat')
vgg16_test_results = loadmat(FILES_PATH + 'VGG16_Test_Index.mat')
densenet121_test_results = loadmat(FILES_PATH + 'Sara_Densenet121_Test_Index_withMask.mat')
snn_test_results = loadmat(FILES_PATH + 'Matching_Test_Index.mat')


labels = pd.read_csv(test_csv)['encoded_labels'].values
inceptionV3 = inceptionV3_test_results['Result']
vgg16 = vgg16_test_results['Result'] 
densenet121 = densenet121_test_results['Result']
snn = snn_test_results['Result']
snn = snn.reshape(snn.shape[1], -1)

densenet121

combined1 = np.concatenate((inceptionV3, vgg16, densenet121), axis = 1)
# combined1 = np.concatenate((vgg16, densenet121), axis = 1)
combined = np.concatenate((combined1, snn), axis = 1)
print(snn.shape, combined1.shape, combined.shape)
combined
# # np.size(concat, 0)
combined = combined.astype(int)

result = []
correct = 0
for i in range(labels.size):
  label = labels[i]
  # print(label)
  freq_label = np.bincount(combined[i]).argmax()
  result.append(freq_label)
  if label == freq_label:
    correct = correct + 1


print(float(correct/labels.size))


# densenet121 = densenet121.astype(int)
# correct = 0
# for i in range(labels.size):
#   label = labels[i]
#   # print(label)
#   d = densenet121[i][0]
#   if label == d:
#     correct = correct + 1


# print(float(correct/labels.size))
```

```python
import sklearn.metrics as sm
scores = sm.precision_recall_fscore_support(np.array(labels),np.array(result) , average='binary')
#print(scores)
precision = scores[0]
recall = scores[1]
fscore = scores[2]
print(fscore)
conf_matrix = sm.confusion_matrix(labels, result)
sensitivity = conf_matrix[0,0]/(conf_matrix[0,0]+conf_matrix[1,0])
specificity = conf_matrix[1][1]/(conf_matrix[0,1]+conf_matrix[1,1])
print(sensitivity)
print(specificity)
```
